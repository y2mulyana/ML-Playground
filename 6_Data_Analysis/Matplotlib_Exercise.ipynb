```python
from PIL import Image
```

```python
im = Image.open('Hausziege_04.jpg')
```

```python
type(im)
```

```python
# Convert image to array
arr = np.array(im)
arr
```

```python
# Berukuran 3600 x 3600 sebanyak 3 layer RGB
arr.shape
```

```python
# plt.imshow(arr)
plt.imshow(arr);
```

```python
plt.imshow(im)
```

```python
gray = arr.mean(axis=2)
gray.shape
```

```python
plt.imshow(gray)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # Creates a 2x2 grid of subplots. axs is a 2x2 array of axes

axs[0, 0].imshow(im)
axs[0, 0].set_title('Original image')

axs[0, 1].imshow(gray)
axs[0, 1].set_title('gray')

axs[1, 0].imshow(gray, cmap='gray')
axs[1, 0].set_title('Grayscale')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# EXERCISE

```python
arr_size = 2000

B = np.random.uniform(low=-1.0, high=1.0, size=(arr_size, arr_size))

colour = np.zeros(arr_size)

for i in range(arr_size):
    if B[0, i] > 0 and B[1, i] < 0:
        colour[i] = 1
    elif B[0, i] < 0 and B[1, i] > 0:
        colour[i] = 1

plt.scatter(B[0], B[1], c=colour);
```
